### Calculating the distance travelled by users

```python
mt_copy["trip_distance"] = np.sqrt(((mt_copy["end_lat"] - mt_copy["start_lat"])**2) + ((mt_copy["end_lon"] - mt_copy["start_lon"])**2))
```

```python
mt_copy["trip_distance"] = mt_copy["trip_distance"] * 111.139
```
